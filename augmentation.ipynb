```python
import csv
import json
import os
import pandas as pd
import cv2
import numpy as np
import albumentations as albu
import glob
from PIL import Image

BACKGROUND_PATHS = glob.glob(os.path.join(".", "background2", "*"))
openImageRow = ["ImageID","LabelName","Confidence","XMin","XMax","YMin","YMax","ClassName"]
files = ["sub-train-annotations-bbox", "sub-validation-annotations-bbox", "sub-test-annotations-bbox"]
file_dirs = [os.path.join(".", "carryber_box")]

new_train_dir = os.path.join(".", "pseudo_data_final_box_big")
os.makedirs(new_train_dir, exist_ok=True)
```

```python
trainrow = []
trainrow.append(openImageRow)
testrow = []
testrow.append(openImageRow)
valrow = []
valrow.append(openImageRow)
```

```python
def cv2pil(image):
    ''' OpenCV型 -> PIL型 ''' 
    new_image = image.copy()
    if new_image.ndim == 2:  # モノクロ
        pass
    elif new_image.shape[2] == 3:  # カラー
        new_image = cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB)
    elif new_image.shape[2] == 4:  # 透過
        new_image = cv2.cvtColor(new_image, cv2.COLOR_BGRA2RGBA)
    new_image = Image.fromarray(new_image)
    return new_image
```

```python
pic_dict = {}
#まず全画像におけるbb情報を取得する
for file_dir in file_dirs:
    for file in files:
        csv_path = os.path.join(file_dir, file+".csv")
        df = pd.read_csv(csv_path)
        image_paths = df["ImageID"].unique()
        for image_path in image_paths:
            image_info = df[df["ImageID"] == image_path]
            if len(image_info) > 1: #だるいので一枚の場合のみいじる
                continue
            else:
                img = cv2.imread(image_path)
                print(image_path)
                h, w = img.shape[:2]  
                for index, row in image_info.iterrows():
                    if ((row["YMax"] - row["YMin"]) * (row["XMax"] - row["XMin"])) > 0.7 or row["XMin"] < 0.02 or row["YMin"] < 0.02 or row["YMax"] > 0.98 or row["XMax"] > 0.98:
                        continue #物体が大きすぎるのでなし
                    pic_dict[image_path] = {
                        "height" : h,
                        "width" : w,
                        "XMin" : w * row["XMin"],
                        "XMax" : w * row["XMax"],
                        "YMin" : h * row["YMin"],
                        "YMax" : h * row["YMax"],
                        "ClassName" : row["ClassName"],
                        "area" : h * w * (row["YMax"] - row["YMin"]) * ( row["XMax"] - row["XMin"])
                    }
```

```python
def left_crop(image, info):
    x1 = int(info["XMin"])
    x2 = int(info["XMax"])
    y1 = int(info["YMin"])
    y2 = int(info["YMax"])
    nsx = int(x1 + np.random.rand() * 0.3 * (x2 - x1))
    h, w = image.shape[:2]
    nw = w - nsx
    nh = int(h * (nw / w))
    info["XMin"] = 0
    info["XMax"] = info["XMax"] - nsx
    if (y1 + y2) / 2 < h / 2:
        new_image = image[0:nh, nsx:]
        info["YMax"] = min(nh, info["YMax"])
    else:
        new_image = image[h-nh:h, nsx:]
        info["YMax"] = info["YMax"] - (h-nh)
        info["YMin"] = max(info["YMin"] -(h-nh) , 0)
    return new_image, info

def right_crop(image, info):
    x1 = int(info["XMin"])
    x2 = int(info["XMax"])
    y1 = int(info["YMin"])
    y2 = int(info["YMax"])
    nfx = int(x2 - np.random.rand() * 0.3 * (x2 - x1))
    h, w = image.shape[:2]
    nw = w - nfx
    nh = int(h * (nw / w))
    info["XMax"] = nfx
    if (y1 + y2) / 2 < h / 2:
        new_image = image[0:nh, 0:nfx]
        info["YMax"] = min(nh, info["YMax"])
    else:
        new_image = image[h-nh:h, 0:nfx]
        info["YMax"] = info["YMax"] - (h-nh)
        info["YMin"] = max(info["YMin"] -(h-nh) , 0)
    return new_image, info


def change_back(image, info):
    h, w = image.shape[:2]
    new_back =  cv2.imread(np.random.choice(BACKGROUND_PATHS))
    new_back = cv2.resize(new_back, (w, h))
    x1 = int(info["XMin"])
    x2 = int(info["XMax"])
    y1 = int(info["YMin"])
    y2 = int(info["YMax"])
    new_back[y1:y2, x1:x2] =  image[y1:y2, x1:x2]
    return new_back, info

def shrink(image, info):
    h, w = image.shape[:2]
    mask_image = image.copy()
    x1 = int(info["XMin"])
    x2 = int(info["XMax"])
    y1 = int(info["YMin"])
    y2 = int(info["YMax"])
    mask_image[y1:y2, x1:x2] = np.zeros(mask_image[y1:y2, x1:x2].shape)
    r = 1 + np.random.rand() * 1.5
    mask_image = cv2.resize(mask_image, (int(w * r), int(h * r)))
    mh, mw  = mask_image.shape[:2]
    sx = int(np.random.rand() * (mw - w))
    sy = int(np.random.rand() * (mh - h))
    mask_image[sy:sy+h, sx:sx+w] = image
    info["XMin"] = info["XMin"] + sx
    info["XMax"] = info["XMax"] + sx
    info["YMin"] = info["YMin"] + sy
    info["YMax"] = info["YMax"] + sy
    return mask_image, info

def show_BB(image, info):
    x1 = int(info["XMin"])
    x2 = int(info["XMax"])
    y1 = int(info["YMin"])
    y2 = int(info["YMax"])
    cv2.rectangle(image, (x1, y1), (x2, y2), (255, 255, 255),  thickness=8)
    display(cv2pil(image))

def zoom(image, info): #boxをでかくすることで処理
    h, w = image.shape[:2]
    x1 = int(info["XMin"])
    x2 = int(info["XMax"])
    y1 = int(info["YMin"])
    y2 = int(info["YMax"])
    target_img = image[y1:y2, x1:x2]
    th, tw = target_img.shape[:2]
    r = 1.5 + np.random.rand()
    target_img = cv2.resize(target_img, (int(tw * r), int(th * r)))
    th, tw = target_img.shape[:2]
    image[y1:y2, x1:x2] = np.zeros(image[y1:y2, x1:x2].shape)
    sx = int(np.random.rand() * max((w - tw), 0))
    sy = int(np.random.rand() * max((h - th), 0))
    image[sy:min(sy+th, h), sx:min(sx+tw, w)] = target_img[:min(th, h), :min(tw, w)]
    info["XMin"] = sx
    info["XMax"] = min(sx+tw, w)
    info["YMin"] = sy
    info["YMax"] = min(sy+th, h)
    return image, info

def random_maker(image, info):
    flag = True
    if np.random.rand() < 1.0:
        try:
            image, info = zoom(image, info)
            flag = False
        except:
            print("error!")
    if (info["XMin"] + info["XMax"]) / 2 < image.shape[1]:
        if np.random.rand() < 0.33:
            try:
                image, info = left_crop(image, info)
                flag = False
            except:
                    print("error!")
    else:
        if np.random.rand() < 0.33:
            try:
                image, info = right_crop(image, info)
                flag = False
            except:
                print("error!")
    """
    if flag and np.random.rand() < 0.5:
        try:
            image, info = shrink(image, info)
        except:
            print("error!")
    """  
    if np.random.rand() < 0.3:
        try:
            image, info = change_back(image, info)
        except:
            print("error!")
        
    #ぼやかす
    if np.random.rand() < 0.15:
        image = cv2.GaussianBlur(image,(3,3),0)
    
    #色をずらす
    if np.random.rand() < 0.15:
        image = albu.augmentations.transforms.RGBShift(p = 1.0)(image=image)['image']
    
    #光度いじる
    if np.random.rand() < 0.10:
        if np.random.rand() < 0.4:
            image = albu.augmentations.transforms.RandomBrightnessContrast(brightness_limit=0.4, contrast_limit=0.5 , p = 1.0)(image=image)['image']
        else:
            image = albu.augmentations.transforms.RandomSunFlare(p = 1.0)(image=image)['image']
    
    return image, info
```

```python
for image_path, info in pic_dict.items():
    print(image_path)
    if not os.path.exists(image_path):
        continue
    image = cv2.imread(image_path)
    r = np.random.rand()
    if r < 0.8:
        for i in range(4):
            new_image_path = os.path.join(new_train_dir, os.path.basename(image_path).split(".")[0] +  "_" + str(i) + ".png")
            #print(new_image_path)
            new_image, new_info = random_maker(image.copy(), info.copy())
            cv2.imwrite(new_image_path, new_image)
            #show_BB(new_image, new_info)
            nh, nw = new_image.shape[:2]
            #print(new_info, nh, nw)
            x1 = new_info["XMin"] / nw
            x2 = new_info["XMax"] / nw
            y1 = new_info["YMin"] / nh
            y2 = new_info["YMax"] / nh
            row = [new_image_path,"takadamaLab_carryber",1.0,x1,x2,y1,y2,new_info["ClassName"]]
            trainrow.append(row)
    elif r < 0.9:
        for i in range(4):
            new_image_path = os.path.join(new_train_dir, os.path.basename(image_path).split(".")[0] +  "_" + str(i) + ".png")
            #print(new_image_path)
            new_image, new_info = random_maker(image.copy(), info.copy())
            nh, nw = new_image.shape[:2]
            cv2.imwrite(new_image_path, new_image)
            #print(new_info)
            x1 = new_info["XMin"] / nw
            x2 = new_info["XMax"] / nw
            y1 = new_info["YMin"] / nh
            y2 = new_info["YMax"] / nh
            row = [new_image_path,"takadamaLab_carryber",1.0,x1,x2,y1,y2,new_info["ClassName"]]
            testrow.append(row)
    else:
        for i in range(4):
            new_image_path = os.path.join(new_train_dir, os.path.basename(image_path).split(".")[0] + "_" +  str(i) + ".png")
            #print(new_image_path)
            new_image, new_info = random_maker(image.copy(), info.copy())
            nh, nw = new_image.shape[:2]
            cv2.imwrite(new_image_path, new_image)
            #print(new_info)
            x1 = new_info["XMin"] / nw
            x2 = new_info["XMax"] / nw
            y1 = new_info["YMin"] / nh
            y2 = new_info["YMax"] / nh
            row = [new_image_path,"takadamaLab_carryber",1.0,x1,x2,y1,y2,new_info["ClassName"]]
            valrow.append(row)
```
